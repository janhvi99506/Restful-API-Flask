## Question 1: What is a RESTful API?

### Theory:
Ans. **RESTful API** (Representational State Transfer API) is a web service that follows the REST architectural style. It enables communication between a **client** and a **server** over HTTP.

**Key principles of REST:**
- **Stateless** – Each request is independent; the server does not store client session data.
- **Client-Server** – The client and server are separated; they interact only through requests/responses.
- **Uniform Interface** – Resources are identified by URLs (e.g., `/users`, `/products`).
- **HTTP Methods** – Uses GET, POST, PUT, DELETE to perform operations.
- **JSON Format** – JSON is the most commonly used format for data exchange.

RESTful APIs are widely used because they are **scalable, simple, and platform-independent**.

````python
# Question 1 - Practical Example
# Install Flask first
!pip install flask

from flask import Flask, jsonify

app = Flask(__name__)

# A RESTful GET endpoint that returns a list of users
@app.route('/users', methods=['GET'])
def get_users():
    return jsonify(['User1', 'User2', 'User3'])

print("RESTful API route defined: GET /users")
print("This endpoint returns a JSON list of users.")
````

## Question 3: What are HTTP methods used in RESTful APIs?

Ans:
HTTP methods define the **type of operation** to be performed on a resource. The main methods used in RESTful APIs are:

| HTTP Method | Purpose | Example |
|-------------|---------|--------|
| **GET** | Retrieve data from server | Get list of users |
| **POST** | Send/Create new data | Add a new user |
| **PUT** | Update/Replace existing data | Update user details |
| **PATCH** | Partially update existing data | Update only user name |
| **DELETE** | Remove a resource | Delete a user |

Each method has a specific purpose, ensuring a **clean and predictable API design**.

````python
# Question 3 - Practical Example
from flask import Flask, jsonify, request

app = Flask(__name__)

# GET - Retrieve data
@app.route('/product', methods=['GET'])
def get_product():
    return jsonify({'name': 'Laptop', 'price': 50000})

# POST - Create new data
@app.route('/product', methods=['POST'])
def add_product():
    return jsonify({'message': 'Product added'}), 201

# PUT - Update data
@app.route('/product/<int:id>', methods=['PUT'])
def update_product(id):
    return jsonify({'message': f'Product {id} updated'})

# DELETE - Remove data
@app.route('/product/<int:id>', methods=['DELETE'])
def delete_product(id):
    return jsonify({'message': f'Product {id} deleted'})

print("All HTTP method routes defined: GET, POST, PUT, DELETE")
````

## Question 4: What is the purpose of the @app.route() decorator in Flask?

Ans:
The **@app.route()** decorator is used to **bind a URL path to a Python function** (called a view function).

**Purpose:**
- It tells Flask **which function to call** when a particular URL is accessed.
- It allows specifying **HTTP methods** (GET, POST, etc.) the route should accept.
- It supports **dynamic URLs** using variables like `<int:id>` or `<string:name>`.
- Without this decorator, Flask would not know how to map incoming requests.

**Syntax:**
```python
@app.route('/path', methods=['GET', 'POST'])
def function_name():
    return 'response'
```

````python
# Question 4 - Practical Example
from flask import Flask

app = Flask(__name__)

# Basic route - GET only (default)
@app.route('/hello')
def hello():
    return 'Hello World'

# Route with specific method
@app.route('/data', methods=['GET', 'POST'])
def data():
    return 'Data endpoint - accepts GET and POST'

# Dynamic route with variable
@app.route('/user/<int:user_id>')
def get_user(user_id):
    return f'User ID: {user_id}'

print("@app.route() decorator examples defined successfully!")
````

## Question 5: What is the role of Flask-SQLAlchemy?

Ans:
**Flask-SQLAlchemy** is an extension that integrates **SQLAlchemy ORM** (Object Relational Mapper) with Flask.

**Key Roles:**
- **Database Connection** – Connects Flask app to databases like SQLite, PostgreSQL, MySQL.
- **ORM Support** – Define database tables as Python classes (models) instead of writing raw SQL.
- **CRUD Operations** – Perform Create, Read, Update, Delete using Python code.
- **Session Management** – Manages database sessions and transactions automatically.
- **Migration Support** – Works with Flask-Migrate for database schema changes.

It makes database interaction **simpler, readable, and maintainable**.

````python
# Question 5 - Practical Example
!pip install flask-sqlalchemy

from flask import Flask
from flask_sqlalchemy import SQLAlchemy

app = Flask(__name__)
app.config['SQLALCHEMY_DATABASE_URI'] = 'sqlite:///school.db'
db = SQLAlchemy(app)

# Define a model (maps to a database table)
class Student(db.Model):
    id   = db.Column(db.Integer, primary_key=True)
    name = db.Column(db.String(100), nullable=False)
    age  = db.Column(db.Integer)

    def __repr__(self):
        return f'<Student {self.name}>'

# Create all tables
with app.app_context():
    db.create_all()
    print("Database and Student table created successfully!")
    print("Table columns: id, name, age")
````

## Question 7: How do you return JSON responses in Flask?

Ans:
In Flask, the **`jsonify()`** function is used to return JSON responses.

**How it works:**
- Converts Python **dictionaries** or **lists** into a proper HTTP JSON response.
- Automatically sets the `Content-Type` header to `application/json`.
- You can also return an **HTTP status code** along with the response.

**Syntax:**
```python
return jsonify({'key': 'value'})         # 200 OK by default
return jsonify({'key': 'value'}), 201    # With custom status code
```

This is the **standard way** to return data from a REST API in Flask.

````python
# Question 7 - Practical Example
from flask import Flask, jsonify

app = Flask(__name__)

# Return a dictionary as JSON
@app.route('/user')
def get_user():
    user = {'name': 'Rahul', 'age': 22, 'city': 'Delhi'}
    return jsonify(user)              # Returns JSON with 200 OK

# Return a list as JSON
@app.route('/items')
def get_items():
    items = ['pen', 'notebook', 'eraser']
    return jsonify(items), 200        # With status code

# Test jsonify directly
with app.app_context():
    with app.test_request_context():
        response = jsonify({'name': 'Rahul', 'age': 22})
        print("JSON Response:", response.get_data(as_text=True))
        print("Content-Type:", response.content_type)
````

## Question 8: How do you handle POST requests in Flask?

Ans:
To handle **POST requests** in Flask:

- The route decorator must include `methods=['POST']`.
- The **`request`** object from Flask is used to access data sent by the client.
- For **JSON data** → use `request.json`
- For **form data** → use `request.form`
- The `get()` method safely retrieves values without raising errors if a key is missing.

**POST requests** are typically used to **create new resources** or send data to the server.

**Syntax:**
```python
@app.route('/endpoint', methods=['POST'])
def handle_post():
    data = request.json
    name = data.get('name')
    return jsonify({'received': name})
```

````python
# Question 8 - Practical Example
from flask import Flask, request, jsonify

app = Flask(__name__)

@app.route('/register', methods=['POST'])
def register():
    data  = request.json              # Parse incoming JSON body
    name  = data.get('name')
    email = data.get('email')

    if not name or not email:
        return jsonify({'error': 'Missing fields'}), 400

    return jsonify({'message': f'{name} registered successfully!'}), 201

# Test POST request using Flask test client
with app.test_client() as client:
    import json
    response = client.post(
        '/register',
        data=json.dumps({'name': 'Priya', 'email': 'priya@example.com'}),
        content_type='application/json'
    )
    print("Status Code:", response.status_code)
    print("Response:", response.get_data(as_text=True))
````

## Question 9: How do you handle errors in Flask (e.g., 404)?

Ans:
Flask provides the **`@app.errorhandler()`** decorator to define **custom error responses**.

**How it works:**
- When an error occurs (e.g., 404), Flask calls the function decorated with `@app.errorhandler(404)`.
- Instead of showing the default HTML error page, you can return a **proper JSON error message**.
- You can handle any HTTP error: **400** (Bad Request), **404** (Not Found), **500** (Server Error), etc.

**Common Error Codes:**
| Code | Meaning |
|------|---------|
| 400 | Bad Request |
| 401 | Unauthorized |
| 403 | Forbidden |
| 404 | Not Found |
| 500 | Internal Server Error |

````python
# Question 9 - Practical Example
from flask import Flask, jsonify

app = Flask(__name__)

@app.route('/home')
def home():
    return jsonify({'message': 'Welcome!'})

# Handle 404 - Not Found
@app.errorhandler(404)
def not_found(error):
    return jsonify({'error': 'Resource not found', 'code': 404}), 404

# Handle 500 - Internal Server Error
@app.errorhandler(500)
def server_error(error):
    return jsonify({'error': 'Internal server error', 'code': 500}), 500

# Test 404 error using test client
with app.test_client() as client:
    response = client.get('/nonexistent-page')
    print("Status Code:", response.status_code)
    print("Response:", response.get_data(as_text=True))
````

Question 10: How do you structure a Flask app using Blueprints?

Ans:
**Blueprints** in Flask allow you to divide a large application into **smaller, reusable modules**.

**Why use Blueprints?**
- Better **code organization** for large applications.
- Each feature (users, products, orders) has its **own separate file**.
- Makes the app more **readable and maintainable**.
- Blueprints are registered on the main app using `app.register_blueprint()`.

**Project Structure with Blueprints:**
```
my_app/
├── app.py          ← Main application file
├── users.py        ← Users Blueprint
├── products.py     ← Products Blueprint
└── orders.py       ← Orders Blueprint
```

````python
# Question 10 - Practical Example
from flask import Flask, Blueprint, jsonify

# ── Blueprint 1: Users ──────────────────────────
users_bp = Blueprint('users', __name__)

@users_bp.route('/users')
def get_users():
    return jsonify(['Alice', 'Bob', 'Charlie'])

@users_bp.route('/users/<int:id>')
def get_user(id):
    return jsonify({'user_id': id, 'name': 'Alice'})

# ── Blueprint 2: Products ───────────────────────
products_bp = Blueprint('products', __name__)

@products_bp.route('/products')
def get_products():
    return jsonify(['Laptop', 'Phone', 'Tablet'])

# ── Main App: Register Blueprints ───────────────
app = Flask(__name__)
app.register_blueprint(users_bp)       # Register users blueprint
app.register_blueprint(products_bp)    # Register products blueprint

# Test both blueprints
with app.test_client() as client:
    r1 = client.get('/users')
    print("Users Response:", r1.get_data(as_text=True))

    r2 = client.get('/products')
    print("Products Response:", r2.get_data(as_text=True))
````
